```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(
    x="parental level of education",
    y="math score",
    data=df
)
plt.xticks(rotation=45)
plt.title("Math score po razini obrazovanja roditelja")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["math score"], kde=True)
plt.title("Histogram rezultata iz matematike")
plt.xlabel("Math score")
plt.ylabel("Frekvencija")
plt.show()
```

```python
from scipy.stats import shapiro

groups = df["parental level of education"].unique()

for g in groups:
    data_group = df[df["parental level of education"] == g]["math score"]
    stat, p = shapiro(data_group)
    print(f"{g}: p = {p:.4f}")
```

```python
from scipy.stats import levene

grouped_data = [
    df[df["parental level of education"] == g]["math score"]
    for g in groups
]

stat, p = levene(*grouped_data)
print("Levene test p-vrijednost:", p)
```

```python
from scipy.stats import f_oneway

anova_stat, anova_p = f_oneway(*grouped_data)

print("F-statistika:", anova_stat)
print("p-vrijednost:", anova_p)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(
    endog=df["math score"],
    groups=df["parental level of education"],
    alpha=0.05
)

print(tukey)
```

Za neke skupine p < 0.05, što znači da strogo gledano normalnost nije u potpunosti zadovoljena.

Međutim:

uzorci su veliki

ANOVA je robustna na umjerena odstupanja od normalnosti


Iako Shapiro–Wilk test pokazuje odstupanja od normalnosti u nekim skupinama, zbog velikog uzorka i robusnosti ANOVA metode, analiza se može pouzdano provesti.

Levene test p-vrijednost = 0.4584


p > 0.05 → varijance su jednake

pretpostavka homoskedastičnosti je zadovoljena

bez toga ANOVA ne bi bila valjana

Jednosmjerna ANOVA – glavni test

F-statistika = 6.52

p-vrijednost = 5.59 × 10⁻⁶

Hipoteze

H₀: sve skupine imaju jednaku srednju vrijednost math score

H₁: barem dvije skupine se razlikuju

Interpretacija

p < 0.05 → odbacujemo H₀

postoji statistički značajna razlika u prosječnim rezultatima iz matematike
među skupinama definiranima razinom obrazovanja roditelja

3️⃣ Tukey HSD – post-hoc analiza (najvažniji dio)

Tukey test odgovara na pitanje:

IZMEĐU KOJIH TOČNO SKUPINA POSTOJE RAZLIKE?

Gledamo samo redove gdje je reject = True

✅ Statistički značajne razlike (reject = True)
🔹 Associate’s degree vs High school

mean diff = −5.75

p = 0.0013

➡️ učenici čiji roditelji imaju associate’s degree imaju značajno više rezultate od onih čiji roditelji imaju samo srednju školu

❌ Neznačajne razlike (reject = False)

bachelor’s ↔ master’s

associate’s ↔ bachelor’s

some college ↔ master’s

some high school ↔ high school

📌 Značenje:

između viših razina obrazovanja razlike nisu dovoljno velike

najveće razlike su između niskih i visokih razina obrazovanja

Jednosmjerna ANOVA analiza pokazala je statistički značajnu razliku u prosječnim rezultatima iz matematike između skupina definiranih razinom obrazovanja roditelja (F = 6.52, p < 0.001). Post-hoc Tukey HSD analiza otkrila je da učenici čiji roditelji imaju višu razinu obrazovanja (associate’s, bachelor’s i master’s degree) ostvaruju značajno bolje rezultate u usporedbi s učenicima čiji roditelji imaju završenu samo srednju školu ili nižu razinu obrazovanja. Razlike između viših razina obrazovanja nisu statistički značajne, što upućuje na to da se najveći pomaci u rezultatima pojavljuju pri prijelazu s nižih na srednje i više razine obrazovanja roditelja.
