```python
import pandas as pd

df = pd.read_csv("/content/sample_data/processed_data2.csv")
print("Oblik skupa podataka:", df.shape)
print("\nVrste podataka (tipovi varijabli):")
print(df.dtypes)

print("\n--- Prvih 10 redaka ---")
display(df.head(10))

print("\n--- Osnovna deskriptivna statistika ---")
display(df.describe().T)

print("\n--- NedostajuÄ‡e vrijednosti po stupcima ---")
display(df.isna().sum()[df.isna().sum() > 0])
```

```python
import numpy as np
import scipy.stats as stats
# Podaci o uzorku
podaci = [22, 25, 20, 23, 24, 19, 21]
n = len(podaci) # VeliÄina uzorka (PAZITI! Manji je od 30)
srednja_vrijednost = np.mean(podaci) # Srednja vrijednost uzorka
standardna_devijacija = np.std(podaci, ddof=1) # Procjena standardne devijacije uzorka (uzorak manji od 30)
# Razina pouzdanosti (npr. 95%)
razina_pouzdanosti = 0.95
alfa = 1 - razina_pouzdanosti
t_vrijednost = stats.t.ppf(1 - alfa / 2, df=n-1) # KritiÄna vrijednost tdistribucije za zadani stupanj slobode
# IzraÄun margine pogreÅ¡ke
margina_pogreske = t_vrijednost * (standardna_devijacija / np.sqrt(n-1))
#PAZITI, FORMULE!
# Interval povjerenja
donja_granica = srednja_vrijednost - margina_pogreske
gornja_granica = srednja_vrijednost + margina_pogreske
print(f"95% interval povjerenja: ({donja_granica:.2f},{gornja_granica:.2f})")
```

S 95% pouzdanosti moÅ¾emo tvrditi da se prava (populacijska) sredina nalazi izmeÄ‘u 19.84 i 24.16.

Shapiroâ€“Wilk test je dobar za male i srednje uzorke (n < 5000).

Kolmogorovâ€“Smirnov test (KS) se koristi opÄ‡enitije, ali zahtijeva standardizaciju podataka.

Ako je p < 0.05, odbacujemo hipotezu normalnosti â†’ podaci nisu normalno distribuirani.

Ako je p â‰¥ 0.05, ne odbacujemo Hâ‚€ â†’ podaci mogu biti normalni.

Prvo Ä‡emo to provjeriti za numeriÄke varijable, recimo price, ram, battery.

```python
import scipy.stats as stats
import numpy as np

# Odaberemo numeriÄke varijable za testiranje
cols_to_test = ['price', 'ram', 'battery']

for col in cols_to_test:
    if col in df.columns:
        x = df[col].dropna()
        print(f"\nðŸ“ˆ Varijabla: {col} (n={len(x)})")

        # Shapiro-Wilk test
        stat, p = stats.shapiro(x)
        print(f" Shapiro-Wilk: W={stat:.4f}, p={p:.4f}")

        # Kolmogorov-Smirnov test (standardizirano)
        z = (x - np.mean(x)) / np.std(x, ddof=1)
        stat2, p2 = stats.kstest(z, 'norm')
        print(f" Kolmogorovâ€“Smirnov: D={stat2:.4f}, p={p2:.4f}")

        # Kratka interpretacija
        if p < 0.05 or p2 < 0.05:
            print(" ðŸ‘‰ Rezultat: Podaci se znaÄajno razlikuju od normalne distribucije.")
        else:
            print(" âœ… Rezultat: Nema znaÄajnog odstupanja od normalne distribucije.")
```

BuduÄ‡i da su p-vrijednosti za obje varijable manje od 0.05, to znaÄi:

Podaci ne prate normalnu distribuciju, pa u daljnjoj analizi ne moÅ¾emo koristiti parametarske testove (poput t-testa ili ANOVA) bez pretvorbi podataka.

Umjesto toga koristimo neparametarske testove, koji ne traÅ¾e pretpostavku normalnosti:

Mann-Whitney U test â†’ za dvije nezavisne skupine,

Kruskal-Wallis â†’ za tri ili viÅ¡e skupina,

Wilcoxon â†’ za parne uzorke.

Ali â€” Äak i ako distribucija nije normalna, Äesto svejedno raÄunamo interval povjerenja za srednju vrijednost ili medianu radi interpretacije.

Zato je sljedeÄ‡i logiÄan korak:

ðŸ”¹ Korak 3 â€” IzraÄun intervala povjerenja (za RAM ili Battery)

```python
import numpy as np
import scipy.stats as stats

x = df['battery'].dropna()

n = len(x)
mean = np.mean(x)
std = np.std(x, ddof=1)
confidence = 0.95
alpha = 1 - confidence


t_value = stats.t.ppf(1 - alpha/2, df=n-1)
margin = t_value * std / np.sqrt(n)

lower = mean - margin
upper = mean + margin

print(f"Srednja vrijednost baterije: {mean:.2f}")
print(f"Standardna devijacija: {std:.2f}")
print(f"95% interval povjerenja: ({lower:.2f}, {upper:.2f})")
```

S 95 % pouzdanosti moÅ¾emo tvrditi da je stvarna prosjeÄna vrijednost kapaciteta baterije u populaciji mobitela izmeÄ‘u 4624 mAh i 4690 mAh.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats


# 2ï¸âƒ£ odaberi numeriÄku varijablu za test (battery)
x = df["battery"].dropna().astype(float)

# 3ï¸âƒ£ osnovne statistike
n = len(x)
mean = np.mean(x)
std = np.std(x, ddof=1)
print(f"VeliÄina uzorka: {n}")
print(f"Srednja vrijednost: {mean:.2f} mAh")
print(f"Standardna devijacija: {std:.2f} mAh")

# 4ï¸âƒ£ postavi hipoteze
# H0: Î¼ = 4500
# H1: Î¼ â‰  4500   (dvostrani test)
mu0 = 4500

# 5ï¸âƒ£ izraÄunaj t-statistiku i p-vrijednost
t_stat, p_value = stats.ttest_1samp(x, popmean=mu0)
print(f"\nRezultat t-testa:")
print(f" t = {t_stat:.4f}")
print(f" p-vrijednost = {p_value:.4e}")

# 6ï¸âƒ£ izraÄun 95% intervala povjerenja
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
margin = t_crit * std / np.sqrt(n)
ci_lower = mean - margin
ci_upper = mean + margin
print(f"\n95% interval povjerenja: ({ci_lower:.2f}, {ci_upper:.2f})")

# 7ï¸âƒ£ izraÄun veliÄine uÄinka (Cohen's d)
cohen_d = (mean - mu0) / std
print(f"Cohen's d = {cohen_d:.3f}")

# 8ï¸âƒ£ odluka o hipotezi
if p_value < alpha:
    print("\nâœ… ZakljuÄak: odbacujemo H0 (prosjeÄna baterija se statistiÄki razlikuje od 4500 mAh).")
else:
    print("\nâ„¹ï¸ ZakljuÄak: ne odbacujemo H0 (nema dovoljno dokaza da se prosjek razlikuje od 4500 mAh).")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import math
import matplotlib.pyplot as plt

# Prikaz grafova unutar Colaba
%matplotlib inline

# 2ï¸âƒ£ UÄitaj podatke (prilagodi put ako su na Driveu)
hormon = pd.read_csv("/content/sample_data/hormon.csv")
knjiga = pd.read_csv("/content/sample_data/knjiga.csv")

print("=== Prvih 5 redaka: hormon.csv ===")
display(hormon.head())

print("\n=== Prvih 5 redaka: knjiga.csv ===")
display(knjiga.head())

# ============================================
# ðŸ”¹ ZADATAK 1: Analiza hormona
# ============================================

# --- 1.1. Interval povjerenja 99% za somatZ ---
hormon_clean = hormon.dropna(subset=['somatZ'])
n = len(hormon_clean)
mean = hormon_clean['somatZ'].mean()
std = hormon_clean['somatZ'].std(ddof=1)
se = std / math.sqrt(n)
alpha = 0.01
tcrit = stats.t.ppf(1 - alpha/2, df=n-1)

ci_lower = mean - tcrit * se
ci_upper = mean + tcrit * se

print("\n--- 1.1. Interval povjerenja za somatZ ---")
print(f"n = {n}")
print(f"Sredina = {mean:.3f}, SD = {std:.3f}")
print(f"99% CI = ({ci_lower:.3f}, {ci_upper:.3f})")

# Provjera normalnosti
shapiro_stat, shapiro_p = stats.shapiro(hormon_clean['somatZ'])
print(f"Shapiroâ€“Wilk test: W = {shapiro_stat:.3f}, p = {shapiro_p:.3f}")
if shapiro_p > 0.05:
    print("âœ… Podaci su pribliÅ¾no normalni â€“ koristi se parametarska metoda (t-interval).")
else:
    print("âš ï¸ Podaci nisu normalni â€“ koristi se neparametarska metoda (npr. bootstrap CI).")
```

```python
male = hormon[hormon['spol'] == 'M']['gastrin'].dropna()
female = hormon[hormon['spol'] == 'Å½']['gastrin'].dropna()

# Provjeri normalnost i varijance
sh_m = stats.shapiro(male)
sh_f = stats.shapiro(female)
lev = stats.levene(male, female)

if sh_m.pvalue > 0.05 and sh_f.pvalue > 0.05:
    # parametarski test
    t_stat, t_p = stats.ttest_ind(male, female, equal_var=(lev.pvalue > 0.05))
    print("\n--- 1.2. t-test (parametarski) ---")
    print(f"t = {t_stat:.3f}, p = {t_p:.3f}")
    if t_p < 0.05:
        print("âœ… Postoji znaÄajna razlika izmeÄ‘u spolova u razini gastrina.")
    else:
        print("â„¹ï¸ Nema znaÄajne razlike izmeÄ‘u spolova.")
else:
    # neparametarski test
    u_stat, u_p = stats.mannwhitneyu(male, female)
    print("\n--- 1.2. Mann-Whitney U test (neparametarski) ---")
    print(f"U = {u_stat:.3f}, p = {u_p:.3f}")
```

```python
smokers = hormon_clean[hormon_clean['pusac'] == 1]['somatZ']
nonsmokers = hormon_clean[hormon_clean['pusac'] == 0]['somatZ']

u_stat_sm, u_p_sm = stats.mannwhitneyu(smokers, nonsmokers)
print("\n--- 1.3. Mann-Whitney test (puÅ¡aÄi vs nepuÅ¡aÄi) ---")
print(f"U = {u_stat_sm:.3f}, p = {u_p_sm:.3f}")
if u_p_sm < 0.05:
    print("âœ… ZnaÄajna razlika izmeÄ‘u puÅ¡aÄa i nepuÅ¡aÄa.")
else:
    print("â„¹ï¸ Nema znaÄajne razlike izmeÄ‘u puÅ¡aÄa i nepuÅ¡aÄa.")
```

```python
ct_spol = pd.crosstab(knjiga['spol'], knjiga['citanje'])
chi_stat1, chi_p1, dof1, exp1 = stats.chi2_contingency(ct_spol)
print("\n--- 2.1. Chi-square test (spol Ã— Äitanje) ---")
print(ct_spol)
print(f"Ï‡Â² = {chi_stat1:.3f}, p = {chi_p1:.3f}")
if chi_p1 < 0.05:
    print("âœ… Postoji znaÄajna povezanost izmeÄ‘u spola i Äitanja.")
else:
    print("â„¹ï¸ Nema znaÄajne povezanosti izmeÄ‘u spola i Äitanja.")
```
