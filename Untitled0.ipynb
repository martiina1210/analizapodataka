```python
import pandas as pd

df = pd.read_csv("/content/sample_data/processed_data2.csv")
print("Oblik skupa podataka:", df.shape)
print("\nVrste podataka (tipovi varijabli):")
print(df.dtypes)

print("\n--- Prvih 10 redaka ---")
display(df.head(10))

print("\n--- Osnovna deskriptivna statistika ---")
display(df.describe().T)

print("\n--- Nedostajuće vrijednosti po stupcima ---")
display(df.isna().sum()[df.isna().sum() > 0])
```

```python
import numpy as np
import scipy.stats as stats
# Podaci o uzorku
podaci = [22, 25, 20, 23, 24, 19, 21]
n = len(podaci) # Veličina uzorka (PAZITI! Manji je od 30)
srednja_vrijednost = np.mean(podaci) # Srednja vrijednost uzorka
standardna_devijacija = np.std(podaci, ddof=1) # Procjena standardne devijacije uzorka (uzorak manji od 30)
# Razina pouzdanosti (npr. 95%)
razina_pouzdanosti = 0.95
alfa = 1 - razina_pouzdanosti
t_vrijednost = stats.t.ppf(1 - alfa / 2, df=n-1) # Kritična vrijednost tdistribucije za zadani stupanj slobode
# Izračun margine pogreške
margina_pogreske = t_vrijednost * (standardna_devijacija / np.sqrt(n-1))
#PAZITI, FORMULE!
# Interval povjerenja
donja_granica = srednja_vrijednost - margina_pogreske
gornja_granica = srednja_vrijednost + margina_pogreske
print(f"95% interval povjerenja: ({donja_granica:.2f},{gornja_granica:.2f})")
```

S 95% pouzdanosti možemo tvrditi da se prava (populacijska) sredina nalazi između 19.84 i 24.16.

Shapiro–Wilk test je dobar za male i srednje uzorke (n < 5000).

Kolmogorov–Smirnov test (KS) se koristi općenitije, ali zahtijeva standardizaciju podataka.

Ako je p < 0.05, odbacujemo hipotezu normalnosti → podaci nisu normalno distribuirani.

Ako je p ≥ 0.05, ne odbacujemo H₀ → podaci mogu biti normalni.

Prvo ćemo to provjeriti za numeričke varijable, recimo price, ram, battery.

```python
import scipy.stats as stats
import numpy as np

# Odaberemo numeričke varijable za testiranje
cols_to_test = ['price', 'ram', 'battery']

for col in cols_to_test:
    if col in df.columns:
        x = df[col].dropna()
        print(f"\n📈 Varijabla: {col} (n={len(x)})")

        # Shapiro-Wilk test
        stat, p = stats.shapiro(x)
        print(f" Shapiro-Wilk: W={stat:.4f}, p={p:.4f}")

        # Kolmogorov-Smirnov test (standardizirano)
        z = (x - np.mean(x)) / np.std(x, ddof=1)
        stat2, p2 = stats.kstest(z, 'norm')
        print(f" Kolmogorov–Smirnov: D={stat2:.4f}, p={p2:.4f}")

        # Kratka interpretacija
        if p < 0.05 or p2 < 0.05:
            print(" 👉 Rezultat: Podaci se značajno razlikuju od normalne distribucije.")
        else:
            print(" ✅ Rezultat: Nema značajnog odstupanja od normalne distribucije.")
```

Budući da su p-vrijednosti za obje varijable manje od 0.05, to znači:

Podaci ne prate normalnu distribuciju, pa u daljnjoj analizi ne možemo koristiti parametarske testove (poput t-testa ili ANOVA) bez pretvorbi podataka.

Umjesto toga koristimo neparametarske testove, koji ne traže pretpostavku normalnosti:

Mann-Whitney U test → za dvije nezavisne skupine,

Kruskal-Wallis → za tri ili više skupina,

Wilcoxon → za parne uzorke.

Ali — čak i ako distribucija nije normalna, često svejedno računamo interval povjerenja za srednju vrijednost ili medianu radi interpretacije.

Zato je sljedeći logičan korak:

🔹 Korak 3 — Izračun intervala povjerenja (za RAM ili Battery)

```python
import numpy as np
import scipy.stats as stats

x = df['battery'].dropna()

n = len(x)
mean = np.mean(x)
std = np.std(x, ddof=1)
confidence = 0.95
alpha = 1 - confidence


t_value = stats.t.ppf(1 - alpha/2, df=n-1)
margin = t_value * std / np.sqrt(n)

lower = mean - margin
upper = mean + margin

print(f"Srednja vrijednost baterije: {mean:.2f}")
print(f"Standardna devijacija: {std:.2f}")
print(f"95% interval povjerenja: ({lower:.2f}, {upper:.2f})")
```

S 95 % pouzdanosti možemo tvrditi da je stvarna prosječna vrijednost kapaciteta baterije u populaciji mobitela između 4624 mAh i 4690 mAh.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats


# 2️⃣ odaberi numeričku varijablu za test (battery)
x = df["battery"].dropna().astype(float)

# 3️⃣ osnovne statistike
n = len(x)
mean = np.mean(x)
std = np.std(x, ddof=1)
print(f"Veličina uzorka: {n}")
print(f"Srednja vrijednost: {mean:.2f} mAh")
print(f"Standardna devijacija: {std:.2f} mAh")

# 4️⃣ postavi hipoteze
# H0: μ = 4500
# H1: μ ≠ 4500   (dvostrani test)
mu0 = 4500

# 5️⃣ izračunaj t-statistiku i p-vrijednost
t_stat, p_value = stats.ttest_1samp(x, popmean=mu0)
print(f"\nRezultat t-testa:")
print(f" t = {t_stat:.4f}")
print(f" p-vrijednost = {p_value:.4e}")

# 6️⃣ izračun 95% intervala povjerenja
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
margin = t_crit * std / np.sqrt(n)
ci_lower = mean - margin
ci_upper = mean + margin
print(f"\n95% interval povjerenja: ({ci_lower:.2f}, {ci_upper:.2f})")

# 7️⃣ izračun veličine učinka (Cohen's d)
cohen_d = (mean - mu0) / std
print(f"Cohen's d = {cohen_d:.3f}")

# 8️⃣ odluka o hipotezi
if p_value < alpha:
    print("\n✅ Zaključak: odbacujemo H0 (prosječna baterija se statistički razlikuje od 4500 mAh).")
else:
    print("\nℹ️ Zaključak: ne odbacujemo H0 (nema dovoljno dokaza da se prosjek razlikuje od 4500 mAh).")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import math
import matplotlib.pyplot as plt

# Prikaz grafova unutar Colaba
%matplotlib inline

# 2️⃣ Učitaj podatke (prilagodi put ako su na Driveu)
hormon = pd.read_csv("/content/sample_data/hormon.csv")
knjiga = pd.read_csv("/content/sample_data/knjiga.csv")

print("=== Prvih 5 redaka: hormon.csv ===")
display(hormon.head())

print("\n=== Prvih 5 redaka: knjiga.csv ===")
display(knjiga.head())

# ============================================
# 🔹 ZADATAK 1: Analiza hormona
# ============================================

# --- 1.1. Interval povjerenja 99% za somatZ ---
hormon_clean = hormon.dropna(subset=['somatZ'])
n = len(hormon_clean)
mean = hormon_clean['somatZ'].mean()
std = hormon_clean['somatZ'].std(ddof=1)
se = std / math.sqrt(n)
alpha = 0.01
tcrit = stats.t.ppf(1 - alpha/2, df=n-1)

ci_lower = mean - tcrit * se
ci_upper = mean + tcrit * se

print("\n--- 1.1. Interval povjerenja za somatZ ---")
print(f"n = {n}")
print(f"Sredina = {mean:.3f}, SD = {std:.3f}")
print(f"99% CI = ({ci_lower:.3f}, {ci_upper:.3f})")

# Provjera normalnosti
shapiro_stat, shapiro_p = stats.shapiro(hormon_clean['somatZ'])
print(f"Shapiro–Wilk test: W = {shapiro_stat:.3f}, p = {shapiro_p:.3f}")
if shapiro_p > 0.05:
    print("✅ Podaci su približno normalni – koristi se parametarska metoda (t-interval).")
else:
    print("⚠️ Podaci nisu normalni – koristi se neparametarska metoda (npr. bootstrap CI).")
```

```python
male = hormon[hormon['spol'] == 'M']['gastrin'].dropna()
female = hormon[hormon['spol'] == 'Ž']['gastrin'].dropna()

# Provjeri normalnost i varijance
sh_m = stats.shapiro(male)
sh_f = stats.shapiro(female)
lev = stats.levene(male, female)

if sh_m.pvalue > 0.05 and sh_f.pvalue > 0.05:
    # parametarski test
    t_stat, t_p = stats.ttest_ind(male, female, equal_var=(lev.pvalue > 0.05))
    print("\n--- 1.2. t-test (parametarski) ---")
    print(f"t = {t_stat:.3f}, p = {t_p:.3f}")
    if t_p < 0.05:
        print("✅ Postoji značajna razlika između spolova u razini gastrina.")
    else:
        print("ℹ️ Nema značajne razlike između spolova.")
else:
    # neparametarski test
    u_stat, u_p = stats.mannwhitneyu(male, female)
    print("\n--- 1.2. Mann-Whitney U test (neparametarski) ---")
    print(f"U = {u_stat:.3f}, p = {u_p:.3f}")
```

```python
smokers = hormon_clean[hormon_clean['pusac'] == 1]['somatZ']
nonsmokers = hormon_clean[hormon_clean['pusac'] == 0]['somatZ']

u_stat_sm, u_p_sm = stats.mannwhitneyu(smokers, nonsmokers)
print("\n--- 1.3. Mann-Whitney test (pušači vs nepušači) ---")
print(f"U = {u_stat_sm:.3f}, p = {u_p_sm:.3f}")
if u_p_sm < 0.05:
    print("✅ Značajna razlika između pušača i nepušača.")
else:
    print("ℹ️ Nema značajne razlike između pušača i nepušača.")
```

```python
ct_spol = pd.crosstab(knjiga['spol'], knjiga['citanje'])
chi_stat1, chi_p1, dof1, exp1 = stats.chi2_contingency(ct_spol)
print("\n--- 2.1. Chi-square test (spol × čitanje) ---")
print(ct_spol)
print(f"χ² = {chi_stat1:.3f}, p = {chi_p1:.3f}")
if chi_p1 < 0.05:
    print("✅ Postoji značajna povezanost između spola i čitanja.")
else:
    print("ℹ️ Nema značajne povezanosti između spola i čitanja.")
```

```python
ct_obr = pd.crosstab(knjiga['obrazovanje'], knjiga['citanje'])
chi_stat2, chi_p2, dof2, exp2 = stats.chi2_contingency(ct_obr)
print("\n--- 2.1. Chi-square test (obrazovanje × čitanje) ---")
print(ct_obr)
print(f"χ² = {chi_stat2:.3f}, p = {chi_p2:.3f}")
```

```python
pct_by_sex = knjiga.groupby('spol')['citanje'].mean() * 100
pct_by_edu_sex = knjiga.groupby(['obrazovanje','spol'])['citanje'].mean().unstack() * 100

print("\n--- 2.2. Postotak čitača po spolu ---")
print(pct_by_sex)
print("\n--- 2.2. Postotak čitača po obrazovanju i spolu ---")
print(pct_by_edu_sex)

# --- Grafovi ---
plt.figure(figsize=(6,4))
pct_by_sex.plot(kind='bar', color=['skyblue','pink'])
plt.title("Postotak čitača po spolu")
plt.ylabel("Postotak (%)")
plt.xlabel("Spol")
plt.show()

plt.figure(figsize=(8,5))
pct_by_edu_sex.plot(kind='bar')
plt.title("Postotak čitača po obrazovanju i spolu")
plt.ylabel("Postotak (%)")
plt.xlabel("Obrazovanje")
plt.legend(title="Spol")
plt.show()

print("\n✅ Analiza završena. Svi rezultati i grafovi prikazani.")
```
