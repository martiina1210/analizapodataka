```python
from scipy.stats import chi2_contingency
import numpy as np

# Ako treba pretvoriti Gender u binarno:
df["Gender_binary"] = df["Gender"].map({"Male":1, "Female":0})

var1 = "Gender_binary"
var2 = "Satisfied"

# ---- Kontingencijska tablica ----
cont = pd.crosstab(df[var1], df[var2])
print("Kontingencijska tablica:\n", cont)

# ---- Phi koeficijent ----
chi2, p, dof, exp = chi2_contingency(cont)
phi = np.sqrt(chi2 / df.shape[0])

print("\nPhi koeficijent =", round(phi, 4))
print("p-vrijednost =", p)
```

```python
from scipy.stats import chi2_contingency
import numpy as np

def cramers_v(conf_matrix):
    chi2, p, dof, exp = chi2_contingency(conf_matrix)
    n = conf_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(conf_matrix.shape)-1)))

cont_table = pd.crosstab(df["Class"], df["Type of Travel"])
print("Kontingencijska tablica:\n", cont_table)

cramersV = cramers_v(cont_table)
print("\nCramer’s V =", round(cramersV, 4))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import numpy as np

df_encoded = pd.get_dummies(df,
                            columns=["Class", "Type of Travel"],
                            drop_first=True)
features = ["Age", "Flight Distance",
            "Departure Delay in Minutes",
            "Arrival Delay in Minutes",
            "Seat comfort", "On-board service"]
X = df_encoded[features]

X_const = sm.add_constant(X)

bool_cols = X_const.select_dtypes(include='bool').columns
X_const[bool_cols] = X_const[bool_cols].astype(int)

X_const = X_const.replace([np.inf, -np.inf], np.nan).dropna()

vif = pd.DataFrame()
vif["Feature"] = X_const.columns
vif["VIF"] = [variance_inflation_factor(X_const.values, i)
              for i in range(X_const.shape[1])]

print(vif)

```

VIF mjeri koliko je svaka varijabla linearno objašnjiva drugim varijablama.

VIF < 5 → super

VIF 5–10 → problem

VIF > 10 → OGROMAN problem → varijablu treba izbaciti

„Na temelju izračuna VIF-a vidimo:

varijabla Departure Delay ima VIF = 12.3 → ekstremna multikolinearnost → treba je ukloniti

varijabla Arrival Delay ima VIF = 11.1 → isto previsoko

sve ostale varijable imaju VIF < 5 → prihvatljive su

Zaključak: model ne smije imati i Departure Delay i Arrival Delay, jer su snažno kolinearne.”


```python
X = df[["Seat comfort",
"Inflight entertainment",
"On-board service",
"Cleanliness",
"Flight Distance"]]
y = df["Satisfied"]
```

```python
import statsmodels.api as sm
X_const = sm.add_constant(X)
logit_model = sm.Logit(y, X_const).fit()
print(logit_model.summary())
```

coef (koeficijent)

Pokazuje kako promena nezavisne varijable utiče na logit (log-odds) verovatnoće da je osoba zadovoljna.

Pozitivan koeficijent → veća verovatnoća zadovoljstva.

Negativan koeficijent → manja verovatnoća zadovoljstva.

std err (standardna greška)

Mera nesigurnosti procene koeficijenta.

z

Z-vrednost = coef / std err → koristi se za testiranje značajnosti.

P>|z|

Ako je < 0.05 → koeficijent je statistički značajan.

[0.025, 0.975]

95% interval poverenja za koeficijent.

Tumačenje pojedinačnih koeficijenata:

const = -5.4218

Osnovna logit verovatnoća kada su sve varijable 0. Vrlo negativno → mala osnovna verovatnoća zadovoljstva.

Seat comfort = 0.0068, p = 0.236

Koeficijent pozitivan, ali nije statistički značajan (p>0.05). Komfort sedišta ne utiče značajno na zadovoljstvo u ovom modelu.

Inflight entertainment = 0.9724, p < 0.001

Veoma značajan i jak pozitivan efekat na zadovoljstvo.

On-board service = 0.4830, p < 0.001

Takođe značajan i pozitivan uticaj.

Čistoća ima pozitivan i značajan efekat, ali slabiji nego zabava ili usluga.

Negativan uticaj, što znači da veće distance blago smanjuju verovatnoću zadovoljstva, iako je efekat po jedinici udaljenosti mali.

```python
params = logit_model.params
odds_ratios = params.apply(lambda b: np.exp(b))
print(odds_ratios)
```

OR = exp(coef)

Odds ratio prevodi logit koeficijent u meru koja je lakša za tumačenje.

Interpretacija: koliko se menjaju šanse (odds) da osoba bude zadovoljna kada se nezavisna varijabla poveća za jednu jedinicu, držeći sve ostale varijable konstantnim.

OR > 1 → povećava verovatnoću događaja (zadovoljstva).

OR < 1 → smanjuje verovatnoću događaja.

OR = 1 → nema efekta.
