```python
from scipy.stats import shapiro

groups = df["parental level of education"].unique()

for g in groups:
    data_group = df[df["parental level of education"] == g]["math score"]
    stat, p = shapiro(data_group)
    print(f"{g}: p = {p:.4f}")
```

```python
from scipy.stats import levene

grouped_data = [
    df[df["parental level of education"] == g]["math score"]
    for g in groups
]

stat, p = levene(*grouped_data)
print("Levene test p-vrijednost:", p)
```

```python
from scipy.stats import f_oneway

anova_stat, anova_p = f_oneway(*grouped_data)

print("F-statistika:", anova_stat)
print("p-vrijednost:", anova_p)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(
    endog=df["math score"],
    groups=df["parental level of education"],
    alpha=0.05
)

print(tukey)
```

Za neke skupine p < 0.05, što znači da strogo gledano normalnost nije u potpunosti zadovoljena.

Međutim:

uzorci su veliki

ANOVA je robustna na umjerena odstupanja od normalnosti


Iako Shapiro–Wilk test pokazuje odstupanja od normalnosti u nekim skupinama, zbog velikog uzorka i robusnosti ANOVA metode, analiza se može pouzdano provesti.

Levene test p-vrijednost = 0.4584

Tukey test odgovara na pitanje:

Post-hoc Tukey HSD analiza otkrila je da učenici čiji roditelji imaju višu razinu obrazovanja (associate’s, bachelor’s i master’s degree) ostvaruju značajno bolje rezultate u usporedbi s učenicima čiji roditelji imaju završenu samo srednju školu ili nižu razinu obrazovanja. 

Prije provođenja Two-Way ANOVA analize izračunate su srednje vrijednosti rezultata iz matematike po kombinacijama faktora spola i pripremnog tečaja. Ovaj korak omogućuje uvid u potencijalne razlike među grupama, ali ne daje informaciju o njihovoj statističkoj značajnosti.

```python
import statsmodels.formula.api as smf

df.columns = df.columns.str.replace(' ', '_')


model = smf.ols(
    "Q('math_score') ~ C(gender) * C(test_preparation_course)",
    data=df
).fit()

print(model.summary())
```

```python
import statsmodels.api as sm

anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

Provedena je dvosmjerna ANOVA kako bi se ispitao utjecaj spola i pohađanja pripremnog tečaja na rezultate iz matematike. Rezultati pokazuju statistički značajan glavni učinak spola (p < 0.05) te pripremnog tečaja (p < 0.05). Interakcijski učinak između spola i pripremnog tečaja nije statistički značajan (p > 0.05), što ukazuje da utjecaj pripremnog tečaja na rezultate iz matematike ne ovisi o spolu učenika.

```python
from statsmodels.multivariate.manova import MANOVA

maov = MANOVA.from_formula(
    'math_score + reading_score + writing_score ~ C(gender) * C(test_preparation_course)',
    data=df
)

print(maov.mv_test())
```

Provedena je multivarijatna analiza varijance (MANOVA) kako bi se ispitao utjecaj spola i pohađanja pripremnog tečaja na rezultate iz matematike, čitanja i pisanja. Rezultati MANOVA analize pokazuju statistički značajan glavni učinak spola (Wilks’ λ = 0.6531, p < 0.05) te statistički značajan glavni učinak pripremnog tečaja (Wilks’ λ = 0.8845, p < 0.05). Interakcijski učinak između spola i pohađanja pripremnog tečaja nije statistički značajan (p = 0.9535), što upućuje na to da učinak pripremnog tečaja ne ovisi o spolu učenika.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# MATEMATIKA
model_math = smf.ols(
    'math_score ~ C(gender) * C(test_preparation_course)',
    data=df
).fit()

anova_math = sm.stats.anova_lm(model_math, typ=2)
print(anova_math)
```

Rezultati univarijantne dvofaktorske ANOVA analize pokazuju da spol i pohađanje pripremnog tečaja imaju statistički značajan utjecaj na rezultate iz matematike. Međutim, njihov interakcijski učinak nije statistički značajan, što ukazuje da učinak pripremnog tečaja na uspjeh ne ovisi o spolu učenika.
